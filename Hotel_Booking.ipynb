# Removing Outlier that can affect further analysis

# adr (Average Daily Rate)

- As (ADR = Number of Rooms Sold/Total Room Revenue )
- So adr cannot be negative, need to remove negative values



```python
negative_values = df[df['adr'] < 0]
negative_values
```

```python
df.head()
```

```python
# A single outlier detected above adr(1000), need to be removed
df = df[df["adr"]<1000]
```

```python
# A single outlier detected which is negative need to be removed
df = df[df["adr"]>0]
```
